## Filtering a DataFrame

```python
ipl.head(2)
```

```python
# find all the final winners
mask = ipl['MatchNumber'] == 'Final'
new_ipl_df = ipl[mask]
new_ipl_df[['Season','WinningTeam']]

ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
ipl.head(2)
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
# how many matches has csk won in kolkata
len(ipl[(ipl['City']=='Kolkata')&(ipl['WinningTeam']=='Chennai Super Kings')])
```

```python
# toss winner is match winner in percentage
(len(ipl[ipl['TossWinner']==ipl['WinningTeam']])/len(ipl))*100
```

```python
# movies with rating higher than 8 and votes>10000
movies.head(2)
movies[(movies['imdb_rating']>8)&(movies['imdb_votes']>10000)].shape[0]
```

```python
# action movies with rating higher than 7.5
len(movies[(movies['genres'].str.split('|').apply(lambda x: 'Action'in x))&(movies['imdb_rating']>7.5)])
```

```python
# wrtie a function that can return the track record of 2 teams against each other
def compare_teams(team1,team2):
    team1wins= len(ipl[ipl['WinningTeam']==team1])
    team2wins= len(ipl[ipl['WinningTeam']==team2])
    return f"{team1} won {team1wins} matches and {team2} won {team2wins} matches."

print(compare_teams('Gujarat Titans','Rajasthan Royals'))
```
